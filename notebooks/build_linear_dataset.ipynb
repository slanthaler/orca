```python
import numpy as np 

input_size = 10
output_size = 1
n_samples = 10_000

#
A = np.random.randn(input_size, output_size)
noise_level = 0.3

#
x = np.random.randn(n_samples, input_size)
y = x @ A + noise_level * np.random.randn(n_samples, output_size)

# split data into train and test
train_size = int(0.8 * n_samples)
x_train, y_train = x[:train_size], y[:train_size]
x_test, y_test = x[train_size:], y[train_size:]

# save to disk
import os
from pathlib import Path
output_dir = Path("../data/linear")
os.makedirs(output_dir, exist_ok=True)
np.savez(output_dir / "linear_train.npz", x_train=x_train, y_train=y_train)
np.savez(output_dir / "linear_test.npz", x_test=x_test, y_test=y_test)
```
